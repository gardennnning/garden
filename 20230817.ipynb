```python
import math
from sklearn import preprocessing                              # 단위를 없애고 0에 맞춰서 분포를 고르게 만들어주기 위함.
from sklearn.preprocessing import MinMaxScaler                 # 단위를 맞춰주는 조치가 scale이다.
from sklearn.preprocessing import StandardScaler
from scipy.stats import boxcox
df_join["총매출액_scale"] = preprocessing.scale(df_join["총매출액"])
df_join.head()
```

```python
scaler = MinMaxScaler()                                   # MinMaxScaler(): 데이터의 스케일을 조정하는 데 사용
x_scale = df_join[['1회 평균매출액','총 할인 금액']]      # scaler하고 싶은걸 꺼내서 해본다.
x_scale_minmax = scaler.fit_transform(x_scale)            # fit_transform로 바꾼다. array로 반환된다.
x_scale_minmax = pd.DataFrame(x_scale_minmax)             # dataframe으로 변환(엑셀형태로 만들어)
x_scale_minmax.head()
```

```python
x_scale_minmax.describe()         # 관측값이 obs=1000 으로 1000개가 잘나왔는지 체크
```

```python
df_join_minmax = pd.concat([df_join,x_scale_minmax], axis=1) #pd가 들어가면 엑셀 형태.
# pd.concat : 여러 개의 데이터프레임을 연결하여 새로운 데이터프레임을 생성 /결합의 함수 (변수의 개수가 맞으면 잘붙는)
# axis=1 열방향으로 붙임
df_join_minmax.head()
```
